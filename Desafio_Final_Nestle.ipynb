```python
#Agora vamos pegar seu conteudo
conteudo = site.content

#agora vamos passar para o beautifulsoup
soup = bs(conteudo, 'lxml')
tabela = soup('table')[0]
tabela
```

```python
#lendo para um df
tabela_ceps = pd.read_html(tabela.prettify(), header=0, flavor='bs4')[0]
tabela_ceps.to_csv('zonas_cep.csv')
```

```python
#verificando tabela
tabela_ceps
```

```python
#Agora que temos um mapeamento de regiões, vamos fazer uma classificação do que tem no nosso dataset
pesquisa['CEP'].fillna(0, inplace=True)
pesquisa['cep_correto'] = pesquisa['CEP'].apply(lambda x: int(x))
```

```python
#Pegando CEP de SP
ceps_sp = pesquisa[pesquisa['MUNICIPIO'] == 'SAO PAULO']
ceps_sp.head()
```

```python
#Criando a ordem do CEP
ceps_sp['CEP_ORD'] = ceps_sp['cep_correto'].apply(lambda x: '0' + str(x)[:2])
```

```python
#Criando ordem no nosso dataset do scrapping
tabela_ceps['CEP_ORD'] = tabela_ceps['Inicio'].apply(lambda x: x[:3])
tabela_ceps.head()
ordem_zonas = tabela_ceps[['Zona', 'CEP_ORD']]
ordem_zonas
```

```python
ceps_sp_zonas = ceps_sp.merge(ordem_zonas, how='left', on='CEP_ORD')
ceps_sp_zonas.drop((ceps_sp_zonas.loc[ceps_sp_zonas['Zona'].isna()].index), inplace = True)
ceps_sp_zonas.shape
```

```python
ceps_sp_zonas.to_excel('sao_paulo_zonas.xlsx')
```

```python
ceps_sp_zonas.columns
```

```python
colunas_modelo = ceps_sp_zonas[['TRABALHADORES',
       'S_RENDA_MEDIA', 'S_POP_TOTAL',
       'S_DOM_PART_PERM', 'S_PCT_CRIANCAS', 'S_PCT_ADOLESCENTE',
       'S_PCT_JOVENS', 'S_PCT_ADULTOS', 'S_PCT_SENIOR', 'S_PCT_IDOSOS',
       'S_PCT_CLASSE_A','S_PCT_CLASSE_B', 'S_PCT_CLASSE_C', 'S_PCT_CLASSE_D', 'S_PCT_CLASSE_E',
       'S_SUPERMERCADO', 'S_TT_PADARIA', 'S_ATACADO', 'S_HIPERMERCADO',
       'S_MINIMERCADO', 'Zona']]

mapa = {
        'Zona Sul': 0,
        'Centro': 1,
        'Zona Leste': 2,
        'Zona Oeste': 3,
        'Zona Norte': 4, 
}

colunas_modelo['Zona'] = colunas_modelo['Zona'].map(mapa)
```

<h1 style = 'font-family: consolas; color: yellow; font-size: 75px'>
            M01
</h1>

---

```python
#Importando bibliotecas de modelos
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

for coluna in colunas_modelo.columns: #realizando a normalização das colunas
    normalizar = MinMaxScaler()
    if coluna != 'Zona':
        colunas_modelo[coluna] = normalizar.fit_transform(np.array(colunas_modelo[coluna]).reshape(-1, 1))
    else:
        continue

#verificando
colunas_modelo.head()
```

```python
#Separando nossos dados
xtrain, xtest, ytrain, ytest = train_test_split(colunas_modelo.iloc[:, :-1], colunas_modelo.iloc[:, -1] , test_size=0.25)

#treinando modelo
floresta = RandomForestClassifier(n_estimators=5)
floresta.fit(xtrain, ytrain)
resultado = floresta.predict(xtest)
df_modelo_prev = pd.DataFrame({'Previsa': resultado, 'Correta':ytest})
df_modelo_prev.head(25)
```

```python
#verificando acertividade do modelo
from sklearn.metrics import confusion_matrix, accuracy_score

print('Matriz de confusão\n', confusion_matrix(ytest, resultado))
print('\nAcurácia do Modelo\n', accuracy_score(ytest, resultado))
```

<h1 style = 'color: #58a4b0'>

```python
#Usando cross validation
from sklearn.model_selection import cross_val_score
floresta_cross = RandomForestClassifier(n_estimators=5, random_state=42)
score_cv = cross_val_score(floresta_cross, colunas_modelo.iloc[: , :-1], colunas_modelo.iloc[:, -1], cv = 5)
score_cv
```

```python
#Média da validação cruzada
print('Acurácia: %0.2f (+/- %0.2f)' %(score_cv.mean(), score_cv.std()*2))
```

```python
#pegando os numeros do Brasil
ceps_geral = pesquisa[pesquisa['MUNICIPIO'] != 'SAO PAULO']
teste_geral = ceps_geral[['TRABALHADORES',
       'S_RENDA_MEDIA', 'S_POP_TOTAL',
       'S_DOM_PART_PERM', 'S_PCT_CRIANCAS', 'S_PCT_ADOLESCENTE',
       'S_PCT_JOVENS', 'S_PCT_ADULTOS', 'S_PCT_SENIOR', 'S_PCT_IDOSOS',
       'S_PCT_CLASSE_A','S_PCT_CLASSE_B', 'S_PCT_CLASSE_C', 'S_PCT_CLASSE_D', 'S_PCT_CLASSE_E',
       'S_SUPERMERCADO', 'S_TT_PADARIA', 'S_ATACADO', 'S_HIPERMERCADO',
       'S_MINIMERCADO']]

for i in teste_geral.columns:
    normalizar = MinMaxScaler()
    teste_geral[i] = normalizar.fit_transform(np.array(teste_geral[i]).reshape(-1, 1))

teste_geral
```

```python
agrupamento_final_pct = pd.DataFrame(ceps_geral[ceps_geral['Previsao_Zona'] == 0].groupby(['UF', 'MUNICIPIO'])['Previsao_Zona'].count()).reset_index()
agrupamento_final_pct.sort_values('Previsao_Zona', ascending=False, inplace = True)
agrupamento_final_pct['Contagem total'] = agrupamento_final_pct['MUNICIPIO'].apply(lambda x: ceps_geral.loc[(ceps_geral['MUNICIPIO'] == x), ['Previsao_Zona']].count())
```

```python
agrupamento_final_pct['porcentagem'] = agrupamento_final_pct['Previsao_Zona']/agrupamento_final_pct['Contagem total']
agrupamento_final_pct.sort_values(['Previsao_Zona', 'porcentagem'], ascending=False, inplace = True)
agrupamento_final_pct['Previsa_Centro_Contragem'] = agrupamento_final_pct['MUNICIPIO'].apply(lambda x: ceps_geral.loc[(ceps_geral['MUNICIPIO'] == x) 
                                                                                                                    & (ceps_geral['Previsao_Zona'] == 1), ['Previsao_Zona']].count())
agrupamento_final_pct.head(20)
```
